# 0.0 Import library

```python
import pandas                       as pd
import numpy                        as np

from matplotlib import pyplot       as plt
from sklearn    import cluster      as ct
from sklearn    import metrics      as mt
```

# 0.1 Load dataset

```python
X = pd.read_csv('../../dataset/clustering/X_dataset.csv')
```

# 1.0 Training model

```python
c = np.arange( 2, 51, 1 )
ss_list = []

for i in c:
    # define model
    kmeans = ct.KMeans( n_clusters=i, max_iter=1000)

    # fit ( training )
    labels = kmeans.fit_predict( X )

    # performance ( test ) 
    ss_avg = np.round(mt.silhouette_score( X, labels ), 3)
    ss_list.append( ss_avg )

max_ss = ss_list.index( max( ss_list ) )
print( 'Para K = {} Avg SS: {}'.format(c[max_ss], max (ss_list) ) )
```

### Graph

```python
plt.plot( c, ss_list)
plt.xlabel( 'Clusters K')
plt.ylabel( 'Silhouette Score Average')
```
