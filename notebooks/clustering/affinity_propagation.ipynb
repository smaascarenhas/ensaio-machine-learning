# 0.0 Import library

```python
import pandas                       as pd
import numpy                        as np

from matplotlib import pyplot       as plt
from sklearn    import cluster      as ct
from sklearn    import metrics      as mt
import time
```

# 0.1 Load dataset

```python
X = pd.read_csv('../../dataset/clustering/X_dataset.csv')
```

# 1.0 Training model

```python
p = np.arange( -1, -51, -1)
ss_list = []

for i in p:
    # model definition
    model = ct.AffinityPropagation( preference=i, max_iter=1000 )
    
    # model training
    model.fit( X )
    
    # model performance
    labels = model.predict( X )

    # metric 
    ss = np.round(mt.silhouette_score( X, labels),3 )
    ss_list.append( ss )

max_ss = ss_list.index( max( ss_list ) )


n_clusters = len(np.unique(labels))
print('preference {} | Avg SS: {} | Número de clusters encontrado: {}'.format(p[max_ss], max(ss_list), n_clusters))
    
```

### Graph

```python
# Graph
plt.plot( p, ss_list)
plt.title( 'Affinity propagation' )
plt.xlabel( 'Preference' )
plt.ylabel( 'Silhouette Score' )
```
